```python
from numpy import *

def Mhat(A):
    evalues,evectors = linalg.eigh(A)
    s = len(A)
    for i in arange(s):
        evalues[i] = round(evalues[i],4)
    j = 0
    Mh = zeros([s,s])
    while j<s-1:
        v = [j]
        while evalues[j] == evalues[j+1]:
            j = j+1
            v.append(j)
        j = j+1
        E = (evectors[:,v]*evectors[:,v].T)
        #print 'E=',E
        Mh = Mh + multiply(E,E)
    if j == s-1:
        E = (evectors[:,j]*evectors[:,j].T)
        #print 'E=',E
        Mh = Mh + multiply(E,E)
    return Mh
def distinct_evalues(A):
    evalues,evectors = linalg.eigh(A)
    s = len(A)
    i = 0
    condition = True
    for i in arange(s):
        evalues[i] = round(evalues[i],4)
    j = 0
    while (condition == True)and(j<s-1):
        if evalues[j]==evalues[(j+1)]:
            condition = False
        j = j+1
    if condition == True:
        return True
    elif condition == False:
        return False
def print_nondistinct(L):
    for i in arange(len(L)):
        B = matrix(L[i].adjacency_matrix())
        H = Graph(L[i].adjacency_matrix())
        distinct = distinct_evalues(B)
        evalues,evectors = linalg.eigh(B)
        if distinct == False:
            H.show()
            hatM = Mhat(B)
            print 'B=',B
            print 'hatM=',hatM
            rankhatM = linalg.matrix_rank(hatM)
            #print evectors
            print 'rank of hatM=',rankhatM
            print 'has distinct eigenvalues:',distinct
            bipartite = H.is_bipartite()
            print 'is bipartite:', bipartite
def print_full_rank(L):
    for i in arange(len(L)):
        B = matrix(L[i].adjacency_matrix())
        H = Graph(L[i].adjacency_matrix())
        hatM = Mhat(B)
        rankhatM = linalg.matrix_rank(hatM)
        if rankhatM == hatM.shape[0]:
            H.show()
            print 'B=',B
            print 'hatM=',hatM
            distinct = distinct_evalues(B)
            print 'has distinct eigenvalues:',distinct
```
