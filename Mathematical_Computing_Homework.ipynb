A series of parts of a homework where we created a cosine-based function with coefficients from our name, then created our own code to calculate the chebyshev approximation of a given function, and tested it out on our name-function (kimmerling) and on the step function which jumps from -1 to 1 at 0.

```python
#Technically the second chunk of the homework but it makes sense to order it here
from numpy import *
from scipy.integrate import quad, quad_vec

def chebcoeff(fct,x,n): 
    #takes in a function, the domain vector, and the desired degree approx, gives chebyshev coefficients for fct approx
    integrand = lambda y: fct(y)*chebyshevT(n,y)/sqrt(1-y*y);
    (prod,err) = quad(integrand,-1,1,epsabs=1e-12);
    return (1+(n>0))*prod/pi;

def chebyshevT(n,x): 
    #x is an array from -1 to 1 with step size h, n is the degree approx we want, gives us the n-th orthogonal T_n(x)
    # Following is a stupid trick to deal with both scalars and ndarrays.
    basis0 = x-x+1.
    if n==0:
        return basis0
    basis1 = x
    if n==1:
        return basis1
    for i in arange(1,n):
        basis2 = 2.*x*basis1-basis0
        basis0 = basis1
        basis1 = basis2
    return basis2

def chebexpansion(fct,K,x): #calculates the Chebyshev expansion of a given function with degree k on a domain vector x
    p = zeros(len(x))
    for k in arange(0,K+1):
        p = p + chebcoeff(fct,x,k)*chebyshevT(k,x)
    return p
```
